```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = x**5
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("pic.jpeg",dpi=100)
```

```python
new_fig = plt.figure()
ax = new_fig.add_axes([0,0,.8,.8])
ax.plot(x,y,label="x,y")
ax.plot(y,x,label="y,x")
ax.legend(loc=(0.1,0.1))
```
