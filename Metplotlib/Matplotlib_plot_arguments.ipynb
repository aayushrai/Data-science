```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = x+3
```

```python
x
```

```python
y
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="red",linewidth=4,alpha=.5,linestyle="--",marker="o",markersize=20,markerfacecolor="yellow",markeredgewidth=4,markeredgecolor="blue")
```

```python
# linewidth or lw argument for width of line
# alpha argument for opacity or we can say tranparency of line
# linestyle example "--","steps","-",":","-." and other check out on website
# marker point mark the point which we take on (x,y) with different symbol exmaple "o","+""
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="red",linewidth=4,alpha=.5,linestyle="--",marker="o",markersize=20,markerfacecolor="yellow",markeredgewidth=4,markeredgecolor="blue")
ax.set_xlim([0,1])   
```
