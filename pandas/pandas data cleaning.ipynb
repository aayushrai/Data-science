# DATA CLEANING

```python
df.isna()
```

```python
df.isna().any()
```

```python
# Removing balance column nan
df["Balance"].fillna(df["Balance"].mean(),inplace=True)
```

```python
#balance column nan is removed
df.isna().any()
```

```python
#removing nan from HasCrCard column
df["HasCrCard"].fillna(df.HasCrCard.value_counts()[1],inplace=True)
```

```python
#HasCrCard column nan is removed
df.isna().any()
```

```python
#removing rownumber coloumn because its not essential
df.drop("RowNumber",axis=1,inplace=True)
```
